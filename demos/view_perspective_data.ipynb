```python
import os
import numpy as np
import matplotlib.pyplot as plt
from astropy.io import fits

# Directory containing perspective images
data_dir = "../perspective_data/perspective_data_linear_fits_radial"

# Get the first 10 .npy files (sorted for consistency)
fits_files = sorted([f for f in os.listdir(data_dir) if f.endswith(".fits")])[:10]

print("Files to display:")
for fname in fits_files:
    print(fname)
```

```python
# Plot the first 10 FITS images
fig, axes = plt.subplots(2, 5, figsize=(20, 8))
axes = axes.flatten()

for i, fname in enumerate(fits_files):
    img = fits.getdata(os.path.join(data_dir, fname))
    ax = axes[i]
    im = ax.imshow(img, origin='lower', cmap='inferno')
    ax.set_title(fname.replace(".fits", ""))
    ax.axis('off')
    fig.colorbar(im, ax=ax, fraction=0.046, pad=0.04)

plt.tight_layout()
plt.show()
```
